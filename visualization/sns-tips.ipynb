```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.day.unique()
```

```python
tips.time.unique()
```

```python
tips['prec'] = tips.tip/tips.total_bill
```

```python
tips.head(3)
```

```python
tips.describe()
```

```python
import numpy as np
g = sns.FacetGrid(tips, col = 'sex', row = 'smoker')
g = g.map(plt.hist, 'prec', bins = np.arange(0,1,10))
```

```python
# fig, ax=plt.subplots(figsize = (11,8)) 
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color ='darkblue', alpha = .2, size = 5)
```

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.lmplot(x = 'total_bill', y= 'tip', data = tips, hue = 'sex', fit_reg = False, palette = 'Set1', )
```

```python

g = sns.FacetGrid(tips, col = 'sex', row = 'time', size = 5)
g = g.map(plt.scatter, 'total_bill', 'tip')
g.savefig('figures/facetgrid_tips.png', dip = 300)
```

```python
g = sns.FacetGrid(tips, col = 'day', hue = 'sex', size = 4)
g = g.map(plt.scatter, 'total_bill', 'tip', edgecolor = 'white').add_legend()
```

```python
tips.pivot_table(values = ['tip','prec'], index = ['sex','smoker'], columns = 'day', margins = True)
```

```python
g = sns.FacetGrid(tips, col= 'sex', row = 'smoker')
g = g.map(plt.scatter, 'total_bill','tip', color = 'g').set(xlabel = 'Bill $', ylabel = '小费')
```
