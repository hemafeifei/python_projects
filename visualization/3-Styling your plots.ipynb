```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reviews = pd.read_csv('csv/winemag-data_first150k.csv', index_col=0)
reviews.head(3)
```

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12,6),
    color='mediumvioletred',
    fontsize=14,
    title='Rankings Given by Wine Magazine')
```
