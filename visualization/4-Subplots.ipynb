```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
reviews = pd.read_csv('csv/winemag-data_first150k.csv', index_col=0)
reviews.head(3)
```

```python
fig, axes = plt.subplots(2,1, figsize=(12,8))
```

```python
axes
```

```python
fig, axes = plt.subplots(2,1, figsize=(12,8))
reviews['points'].value_counts().sort_index().plot.bar(ax=axes[0])
reviews['province'].value_counts().head(15).plot.bar(ax=axes[1])
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))

reviews['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0][0]
)

reviews['province'].value_counts().head(20).plot.bar(
    ax=axarr[1][1]
)
```

```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(2,2, figsize=(12, 8))
reviews['points'].value_counts().sort_index().plot.bar(
    ax=axes[0][0], fontsize=12, color='mediumvioletred')
axes[0][0].set_title("Wine Scores", fontsize=18)

reviews['variety'].value_counts().head(15).plot.bar(
    ax=axes[0][1], fontsize=12, color='mediumvioletred')
axes[0][1].set_title("Wine Variety", fontsize=18)

reviews['province'].value_counts().head(15).plot.bar(
    ax=axes[1][1], fontsize=12, color='mediumvioletred')
axes[1][1].set_title("Wine Origins", fontsize=18)

reviews['price'].value_counts().plot.hist(
    ax=axes[1][0], fontsize=12, color='mediumvioletred')
axes[1][0].set_title("Wine Prices", fontsize=18)

plt.subplots_adjust(hspace=.3)
sns.despine()
```
