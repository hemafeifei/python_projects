```python
def parse_cb_data(soup):
    table = []
    for tr in soup.find_all('tr')[2:]:
        lst = []
        for td in tr.find_all("td"):
            lst.append(td.get_text())
        table.append(lst)
    return table
```

```python
soup_cb = get_soup('https://www.jisilu.cn/data/cbnew/#cb')
```

```python
table = parse_cb_data(soup_cb)
```
