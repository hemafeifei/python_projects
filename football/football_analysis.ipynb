```python
import pandas as pd
```

```python
df_es1 = pd.read_csv('data/results_Spanish_A_2009-2017.txt',encoding = 'gbk')
df_es1.overtime.unique()
```

```python
df_es1.shape
```

```python
df_es1 = df_es1.loc[df_es1.overtime!='取消']
df_es1.shape
```

```python
df_es1['over_goals'] = pd.to_numeric(df_es1['overtime'].str.split(':').str[0]) + pd.to_numeric(df_es1['overtime'].str.split(':').str[1])
df_es1.head()
```

```python
pd.pivot_table(df_es1[['half','overtime','over_goals']], index = 'half', columns = 'over_goals', aggfunc='count')
```

```python
pd.pivot_table(df_es1[['half','overtime','over_goals']], index = 'half', columns = 'over_goals', aggfunc='count').to_csv('ES_A.txt',encoding ='gbk')
```
