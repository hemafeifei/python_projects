```python
half_goals = []
over_goals = []
for x in range(len(df_b1)):
    half_goals.append(pd.to_numeric(df_b1['half'][x][0]) + pd.to_numeric(df_b1['half'][x][2]))
    over_goals.append(pd.to_numeric(df_b1['overtime'][x][0]) + pd.to_numeric(df_b1['overtime'][x][2]))
```

```python
df_b1['half_goals'] = half_goals
df_b1['over_goals'] = over_goals
df_b1.head()
```
