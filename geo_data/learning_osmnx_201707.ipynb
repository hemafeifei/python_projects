### 修改路网 

```python
# street network from lat_long point
G1 = ox.graph_from_point((23.1255, 113.3751), distance =1250, network_type='drive_service' )
ox.plot_graph(G1)
```

```python
#找出不是端点的节点
nc = ['b' if ox.is_endpoint(G1, node) else 'r' for node in G1.nodes()]
fig, ax = ox.plot_graph(G1, node_color=nc, node_zorder=3)
```

```python
#simplify the network
# G1= ox.simplify_graph(G1)
fig, ax = ox.plot_graph(G1, node_color='w', node_edgecolor='k', node_zorder=3)
```

```python
#highligh all the one-way edges
ec = ['r' if data['oneway'] else 'b' for u,v,key,data in G1.edges(keys = True, data=True)]
fig, ax =ox.plot_graph(G1, node_size=0, edge_color=ec, edge_linewidth=1.5, edge_alpha=0.5)
```

```python
def edge_color():
    for u,v,key,data in G1.edges(keys = True, data=True):
        if data['highway'] == 'residential':
            return 'r'
#         elif data['highway'] == 'primary':
#             return 'm'

ec = ['r' if data['highway'] == 'residential' else 'b' for u,v,key,data in G1.edges(keys = True, data=True)]
# G1.edges(keys=True,data=True)[3]
fig, ax =ox.plot_graph(G1, node_size=0, edge_color=ec, edge_linewidth=1.5, edge_alpha=0.5)
```

```python
#修剪节点
import numpy as np
import matplotlib.pyplot as plt
intersections = ox.clean_intersections(ox.project_graph(G1), tolerance=20, dead_ends=False)
points = np.array([point.xy for point in intersections])
```

```python
points[0]
```

```python
fig, ax = ox.plot_graph(ox.project_graph(G1), fig_height=10, show=False, close=False, node_alpha=0)
ax.scatter(x=points[:,0], y = points[:,1], zorder=2, color = '#66ccff', edgecolors='k')
plt.show()
```

```python
#查询路网的属性
stats = ox.basic_stats(G1)
stats['circuity_avg']
# extented_stats = ox.extented_stats(G, ecc =True, bc=True, cc =True)
#extended_stas也是一个字典，可以通过k,v合并到stats中
```

### 分析路网 

```python
#找最短路径
import networkx as nx
desti = list(G1.nodes())[0]
origin = list(G1.nodes())[-1]
route = nx.shortest_path(G6,origin, desti)
fig, ax = ox.plot_graph_route(G6,route)
```

```python
#两个不在路网节点中的点
# Latitude: 23.1292,Longitude: 113.3726
# Latitude: 23.1232,Longitude: 113.3641
origin_point = (23.1292, 113.3726)
desti_point =( 23.1232,113.3641)
#寻找离两个点最近的node
origin_node = ox.get_nearest_node(G1, origin_point)
desti_node = ox.get_nearest_node(G1, desti_point)
```

```python
#make route
route = nx.shortest_path(G1, origin_node, desti_node, weight='length')
str(route)
```

```python
#plot the route and two point away from node
fig, ax = ox.plot_graph_route(G1, route, origin_point=origin_point, destination_point= desti_point)
```
